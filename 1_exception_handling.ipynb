```python
#Try , except and else block
try:
    num = int(input("Enter a number:"))
    result = 10/num

except ValueError as ve:
    print("Please enter a valid number")
except ZeroDivisionError as zde:
    print("Division by zero is not allowed")

except Exception as e:
    print("An unexpected error occurred:", e)

else:
    print("The result is:", result)

#Else will be excecuted only if there is no exception in try block
```

```python
#Try, except, else and finally block
try:
    num = int(input("Enter a number:"))
    result = 10/num

except ValueError as ve:
    print("Please enter a valid number")
except ZeroDivisionError as zde:
    print("Division by zero is not allowed")

except Exception as e:
    print("An unexpected error occurred:", e)

else:
    print("The result is:", result)

finally:
    print("Execution completed.")

#Finally block will always be excecuted regardless of whether an exception occurred or not.
#This is good practise when exception ocure in database connection or file handling where you want to close the connection or file in any case.
```
