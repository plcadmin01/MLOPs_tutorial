Generator is a simpler way to create itertors. They use yield keyword to produce a series of values lazily
meaning generate the value in fly and do not store them in memory

```python
def square(n):
    for i in range(n):
        return i**2
```

```python
square(3)
```

```python
def square(n):
    for i in range(n):
        yield i**2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a = square(3)
```

```python
next(a)
```

```python
#practical example
#Reading large files
```

```python
def read_large_file(file_name):
    with open(file_name) as f:
        for line in f:
            yield line
            
```

```python
file_name = 'large_file.txt'

for line in read_large_file(file_name):
    print(line.strip())
```
