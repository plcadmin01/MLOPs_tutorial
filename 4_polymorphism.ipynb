Polymorphism is a core concept in OOP, where objects of dofferent classes to be treated as object of common superclass
Polymorphism is typically achieved through method overriding and interfaces

single action in different form (poly- multiple forms)

```python
#Base Class
class Animal:
    def speak(self):
        return "Sound of an animal"
```

```python
#Derived class
class Dog(Animal):
    def speak(self):
        return "Woof Woof"
```

```python
#Derived class 2
class Cat(Animal):
    def speak(self):
        return "Meow Meow"
```

```python
dog1 = Dog()
```

```python
dog1.speak()
```

```python
cat1 = Cat()
cat1.speak()
```

```python
def animal_sound(animal):
    print(animal.speak())
```

```python
animal_sound(dog1)
```
