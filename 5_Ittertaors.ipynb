Itterators are advance python concepts which allow efficient looping and memory management.
It is a way to access elements of collection sequentially without exposing underlying struture

```python
my_list = [1, 2, 3, 4, 5]
```

```python
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
#Iterator, using Iter keyword
my_iter = iter(my_list)
```

```python
print(type(my_iter))
```

```python
my_iter
```

```python
#It uses lazy evaluation
next(my_iter)  #1

```

```python
try:
    next(my_iter)  #2

except StopIteration:
    print("No elements in the iterator")
```
