```python
#decorator
def main_welcome(func):
    msg = "Welcome"

    def sub_welcome():
        print("Welcome to the Clinic")
        func()
        print("Hello!!!")
        

    return sub_welcome()
```

```python
def course_intro():
    print("This is a ML Ops Tutorial")
```

```python
course_intro()
```

```python
main_welcome(course_intro)
```

```python
@main_welcome
def course_intro():
    print("This is a ML Ops Tutorial")
```

```python
#Example 2 of decorator

#I have a closure function
def decorator_function(func):
    def wrapper_function():
        print("This is before executing the function")
        func()
        print("This is after executing the function")
    return wrapper_function
```

```python
@decorator_function
def say_hello():
    print("Hello!")
```

```python
#Every decorator is a closure function but every closure function is not a decorator function 
#âˆšA closure is a nested function that "remembers" a variable from its parent function, even after the parent function has finished executing.
```

```python
def make_multiplier(x):
    def multiplier(n):
        return x * n  # It "remembers" what 'x' was
    return multiplier

times3 = make_multiplier(3)
print(times3(10))  # Output: 30
```

```python
#A decorator is a closure that specifically takes a function as its input and returns a modified version of it.
def my_decorator(func):
    def wrapper():
        print("Something is happening before.")
        func()
        print("Something is happening after.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

```

```python
say_hello()
```

Main Purpose	To "save" a value or state for later use.	To add features to an existing function.
Input	Usually data (integers, strings, etc.).	Always a function.
Syntax	Standard function calls.	Uses the @decorator_name shortcut.
Analogy	A backpack that holds specific items for you to use later.	A gift wrap that adds a bow and a card to a present.

```python
#Decorator with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator
```

```python
@repeat(3)
def greet(message):
    print(message)
```

```python
greet("hello")
```
